```
##1>What is the average median_income of the data set and check the distribution of data using appropriate plots?
##please explain the distribution of plots.
##1>average median_income of data
average=df["median_income"].mean()
average
```

```
##1>distribution of data using plot
sns.histplot(df.median_income,color='g')
plt.show()
##in this median income plot here the data distribution is highly skewed
```

```
##3>show with the help of visualization,how median_income and median_house_values are releted?
##3>median_income vs median_house_value,here both are numerical data so we are going for scatterplot
plt.scatter(df['median_income'],df['median_house_value'],color='g',marker='x')
plt.title('median_income vs median_house_value') ##title of the scatter plot
plt.xlabel('median_income') ## giving x axis name
plt.ylabel('median_house_value')
plt.show()
##from this graph of median house value and median income we can understand that the points are intact and are increasing
```
